```python
import re
import json
import requests
import bs4
import urllib.request
```

```python
session = requests.session()
```

```python
videoId = "141724"
responseHTML = session.get("http://tvple.com/" + videoId)
```

```python
parseSoup = bs4.BeautifulSoup(responseHTML.text, 'html5lib')
```

```python
pattern = re.compile(r'http://api.tvple.com/v1/player/meta\?data=')
pattern
```

```python
dataMeta = parseSoup.find(True, {"data-meta": pattern})
dataMetaUrl = dataMeta['data-meta']
dataMetaUrl
```

```python
infoJsonData = session.get(dataMetaUrl)
infoJsonData.text
```

```python
jsonDicData = json.loads(infoJsonData.text)
print(jsonDicData)
```

* 자막 링크 가져오기

```python
cloudReadUrl = jsonDicData["cloud"]["read_url"]
cloudReadUrl
```

* 영상 데이터 링크 가져오기

```python
videoUrl = jsonDicData["stream"]["sources"]["a"]["urls"]["mp4_avc"]
videoUrl
```

```python
cloudData = session.get(cloudReadUrl)
cloudData
```

```python
cloudJsonData = json.loads(cloudData.text)
cloudJsonData["0"]
```

```python
urllib.request.urlretrieve(videoUrl, videoId + ".mp4")
```

```python
with open(videoId + '.json', 'w') as outfile:
    json.dump(cloudJsonData, outfile)
```
